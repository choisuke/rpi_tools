# プログラムの概要  
* その時点におけるアナログメーターの値を読み取る

## インポート

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## 検出する画像

```python
tmp = cv2.imread('certain_time_cpu.jpg')
tmp = cv2.cvtColor(tmp, cv2.COLOR_RGB2BGR)
plt.imshow(tmp)
```

## 差分画像の作成

```python
#元画像の読み込み
criterion_src = cv2.imread('criterion_cpu.jpg', 0) #グレースケールで読み込み
certain_time_src = cv2.imread('certain_time_cpu.jpg', 0) #グレースケールで読み込み

#差分画像の作成して、閾値以上は255にする
diff_img = cv2.absdiff(certain_time_src, criterion_src)
diff_img[diff_img > 10] = 255
#diff_img[diff_img <= 10] = 0

plt.imshow(diff_img, cmap='gray', vmin=0, vmax=255)
```

```python
#ノイズ除去をする
kernel = np.ones((3,3),np.uint8)
#オープニング処理（収縮1回→膨張1回）
diff_img_opening = cv2.morphologyEx(diff_img, cv2.MORPH_OPEN, kernel)

plt.imshow(diff_img_opening, cmap='gray', vmin=0, vmax=255)
```

## 値を読み取る処理の作成

```python
#読み取った値を格納用
value = 0
#1度当たりのvalue
value_per_angle = 0.4347826
#計算する範囲（値が0,max,針の中心の3点のなす角を計算。場合によっては360度からマイナスする）
max_angle = 230
#判定用の画像
judge_img = diff_img_opening
#画像を回転するときの中心座標
center = (106, 120) #(横、縦)

#指定した角度の範囲で回転する角度を1度づつ増やして、針の位置を検出。
for angle in range(1, max_angle+1):
    judge_img_crop = judge_img[146:150, 40:43] #目視で確認した位置[縦, 横]
    
    #judge_img_cropが白色なら針がその位置にある取るする
    if np.count_nonzero(judge_img_crop > 200) > 9: #cropした画像の中に200以上のピクセルが9個より多かったらその位置にあるとする
        value = angle * value_per_angle
        break
    '''
    #仮
    if angle == 20:
        tmp = judge_img
        tmp_crp = judge_img_crop
    '''
    #針を検出できなかった場合はjudge_imgを回転させる
    trans = cv2.getRotationMatrix2D(center, angle , 1)
    judge_img = cv2.warpAffine(diff_img_opening, trans, (judge_img.shape[1],judge_img.shape[0]))
```

```python
print('針の位置（角度）：' + str(angle))
print('現時点での値：' + str(value))
```

## どこをトリミングしているのか確認

```python
plt.imshow(judge_img, cmap='gray', vmin=0, vmax=255)
```

```python
#カラーに変換
judge_img_rgb = cv2.cvtColor(judge_img, cv2.COLOR_GRAY2BGR)
```

```python
judge_img_rigion = cv2.rectangle(judge_img_rgb, (40,146 ), (43, 140), (0, 0, 255), 1)
```

```python
cv2.imwrite('judge_img_rigion.jpg', judge_img_rigion)
```
